Observations
* For some features (e.g. `area_mean`, `area_worst`), there are differences between the mean and median, which provides initial indications of non-normal distributions.

```python
# display descriptive statistics for the numeric features for each of the two tumor groups
df.groupby("diagnosis").agg(['mean', 'median'])
```

Observations
* For all `_mean` features except `fractal_dimension_mean`, the mean and median of the features in the group of benign tumors are lower than in the group of malignant tumors.
* For the `_se` features, this difference is not present for all features. However, it is particularly strong for the features `radius_se`, `perimeter_se` and `area_se`. This indicates that there is a greater variance in the size of the cell nuclei in malignant tumors than in benign tumors.
* The differences are even more pronounced in the `_worst` features than in the `_mean` features. Here too, the differences between the two groups are particularly large for the `radius`, `perimeter` and `area` features. This means that not only the variance in the size of the cell nuclei is higher in malignant tumors than in benign tumors, but also the overall size of the cell nuclei. 

```python
# evaluate data skewness and kurtosis
skewness = df_float.skew()
kurtosis = df_float.kurt()
df_skew = pd.DataFrame({"skewness": skewness, "kurtosis": kurtosis})
df_skew
```

Observations
* Based on the values of skewness and kurtosis, the distribution of `smoothness_worst`, `concave_point_worst`, `texture_worst` and `smoothness_mean` can be described as “normal”. 
* The `_se` features in particular deviate strongly from the normal distribution, but the `_mean` and `_worst` features also provide indications of non-normally distributed data.

### 3.1 Correlation Analysis

```python
# analyze (non-parametric) correlations between numerical features
corr_matrix = df_float.corr(method="kendall") 

# create a heatmap that visualizes the correlation sizes
plt.figure(figsize=(16, 16))
cmap = sns.diverging_palette(220, 300, s=60, center="light", as_cmap=True)
sns.heatmap(corr_matrix, cmap=cmap, center=0)
plt.show()
```

Observations
* Most correlations between the features are positive. Only a few are negative. This means that higher values in one feature are associated with higher values in other features. 
* In particular, the three features `radius`, `perimeter` and `area` correlate very strongly with each other.   
* Many correlations are rather low to moderate (|r| < 0.4; Cohen, 1988, pp. 79-80), i.e. the different features measure different aspects. 

```python
# analyze correlation sizes further
dp.visualize_correlation_sizes(corr_matrix, absolute_values=False)
```

Observations
* Most correlations are in the range between 0.2 and 0.3.
* 28% of the correlations are greater than 0.4.
* 17% of the correlations are greater than 0.5.

```python
# analyze interaction / moderation effects by calculating the difference in the correlation between two numeric features in the two tumor groups
corr_matrix_diff = dp.calculate_corr_diffs(df)

plt.figure(figsize=(16, 16))
cmap = sns.diverging_palette(220, 300, s=60, center="light", as_cmap=True)
sns.heatmap(corr_matrix_diff, cmap=cmap, center=0)
plt.show()
```

Observations
* The correlation of some features differs greatly depending on the group to which they belong.
* For example, the correlation between `radius_se` and `radius_mean` is .5 higher in the malignant tumor group than in the benign tumor group. 
* For features where the correlation differs strongly depending on the group membership, it may make sense to include interaction terms. 
* The features usually correlate more positively (purple) in the malignant tumor group than in the benign tumor group.

```python
# analyze correlation differences further
dp.visualize_correlation_sizes(corr_matrix_diff, differences=True)
```

Observations
* There are no major group differences in the majority of correlations.
* In 86% of the correlations, the group difference is less than 0.2.
* Only in 6% of the correlations is the group difference greater than 0.3. 

```python
# generate a DataFrame containing only the features with high correlation differences between the two groups
filtered_corr = dp.filter_corr(corr_matrix_diff)

# visualize these correlations as scatter plots
num_plots = len(filtered_corr)
cols = 3
rows = (num_plots + cols - 1) // cols  # calculate the number of rows

# create the scatter plots
fig, axes = plt.subplots(rows, cols, figsize=(15, 5 * rows))
axes = axes.flatten()

for ax, (index, row) in zip(axes, filtered_corr.iterrows()):
    feature1 = row["Feature1"]
    feature2 = row["Feature2"]

    sns.scatterplot(data=df, x=feature1, y=feature2, hue="diagnosis", palette=palette, ax=ax)
    ax.set_xlabel(feature1)
    ax.set_ylabel(feature2)
    ax.grid(True)

# delete unused subplots
for i in range(len(filtered_corr), len(axes)):
    fig.delaxes(axes[i])

# adjust space between rows
plt.subplots_adjust(hspace=0.4)

plt.tight_layout()
plt.show()
```

Observations
* The scatterplots clearly show the two clusters of the two diagnoses. While malignant tumors are often characterized by higher values for both features, benign tumors usually have lower values, even if there is overlap between the two groups in the lower value range.
* Isolated points are visible that are clearly removed from the bulk of the data. These may be potential outliers. They are exclusively assigned to the malignant tumor group.
* Only the features `perimeter`, `area` and `radius` are shown in the diagrams. This means that for these features, there are differences between the two tumor groups, particularly in the interaction of these features and their characteristics with each other.
* The features are exclusively features that describe the size of the cell nuclei.
* For each correlation, a `_worst` / `_mean` feature is correlated with a `_se` feature.
* The scatter plots show that the correlations are usually lower in the group of benign tumors than in the group of malignant tumors. 

### 3.2 Outlier Identification

```python
# determine the number of columns and rows for the plot
n_rows = 8
n_cols = 4

# create the subplots 
fig, axes = plt.subplots(n_rows, n_cols, figsize=(12, 30))
axes = axes.flatten()

# create a boxplot for each feature
boxprops = dict(linestyle='-', linewidth=2)
for i, col in enumerate(df_float.columns[:-1]):  
    sns.boxplot(y=col, data=df, ax=axes[i], boxprops=boxprops)

# delete empty subplots
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

Observations 
* When using the interquartile range (IQR) proximity rule (Galli, 2022, p. 154) to identify outliers, almost every feature (except `concave points_worst`) results in outliers.
* Almost all outliers are located in the upper value range of the features. 

```python
# generate a new DataFrame, where outliers in each column are indicated
df_out = dp.calculate_outliers(df_float)

# get outlier descriptives
display(dp.get_outlier_descriptives(df_out).describe())

# visualize outliers
dp.visualize_outliers(df_out)
```

Observations
* For the majority of records (70%), no values were marked as outliers. One value was identified as an outlier for 10% of records. For approx. 6% of the records, more than 5 values were identified as outliers.
* All columns have at least 5 outliers, except the `concave points_worst` column, where no value was identified as an outlier.
* On average, 3.4% of the values in the columns were identified as outliers. 
* The `area_se` feature has by far the most outliers (11.4% of the values). 
* More values were identified as outliers in the `_se` features than in the other features. 

```python
# create a copy of the outlier DataFrame
df_out_dia = df_out.copy()

# calculate the sum of outliers for each row
outlier_sum = df_out.sum(axis=1)

# add a new column to indicate whether a row has at least one outlier
df_out_dia["is_outlier"] = [1 if val > 0 else 0 for val in outlier_sum]

# copy the diagnosis column from the original DataFrame
df_out_dia["diagnosis"] = df["diagnosis"].copy()

# group the DataFrame by is_outlier and diagnosis and count the number of rows in each group
df_out_dia.groupby(["is_outlier", "diagnosis"]).size()
```

Observations
* Malignant tumors were identified as outliers in more features than benign tumors, although malignant tumors are less common in the dataset than benign tumors.
